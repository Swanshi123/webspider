```python
## Step1 Reading and cleaning text
import pandas as pd
df=pd.read_excel('./7005 990评论.xlsx')
df
```

```python
df_text=df[['nickname','content']]
df_text
df_text['words'] = df_text["content"].apply(lambda x: x.split(" "))
df_text
```

```python
ls = ["nickname", "words"]
df2 = df_text[ls].reset_index(drop=True)
df2.tail(2)
df2= df2.explode("words")
df2
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
%config InlineBackend.figure_format = 'retina'## This is to view a high-resolution figure in Jupyter Notebook
```

```python
## This is a list of all words
text = df2["words"].tolist()
## Join words using " " into a string. " " will not affect the results. Just put words together.
text = " ".join(text)
```

```python
text
```

```python
## Step 2 uses the jieba segmentation package for word segmentation
import jieba
jieba.load_userdict(r'../baidu_stopwords.txt')
words = jieba.lcut(text,cut_all=False)
```

```python
# Creating word_cloud with text as argument in .generate() method
word_cloud = WordCloud(font_path="/System/Library/Fonts/Hiragino Sans GB.ttc",width = 1200, height = 800,
                       collocations = False, 
                       background_color = 'white',
                       min_font_size = 5, max_font_size = 80).generate(text)
##For Chinese, add the font path

## This is the size
fig = plt.figure(figsize = (12, 8))

## This is to plot the figure
plt.imshow(word_cloud, interpolation = 'hamming')
plt.axis('off')

# Save the figure
fig.savefig('Fig8.pdf', format='pdf', dpi=300, bbox_inches='tight')
```

```python
## Customize word clouds for specific graphics
## Load the libraries
from wordcloud import WordCloud, ImageColorGenerator
from PIL import Image
import numpy as np
```

```python
## Let's try to do a "Heart" layout!!

## Read the mask image
mask1 = np.array(Image.open("../bear1.jpg"))
plt.imshow(mask1)
plt.axis('off')
```

```python
word_cloud1 = WordCloud(font_path="/System/Library/Fonts/Hiragino Sans GB.ttc",background_color = 'white', mask = mask1, 
               width = 600, height = 600,
                      min_font_size = 15, max_font_size = 55).generate(text)

## This is the size
fig = plt.figure(figsize = (8, 8))

# font color matching the masked image
img_colors = ImageColorGenerator(mask1)
word_cloud1.recolor(color_func = img_colors)

## This is to plot the figure
plt.imshow(word_cloud1, interpolation = 'hamming')
plt.axis('off')

# Save this figure
#fig.savefig('Figure/Fig9.pdf', format='pdf', dpi=200, bbox_inches='tight')
```
