```python
import requests
import json
import pandas as pd
import time
```

```python
items = []
for i in range(1,10):
    time.sleep(3) 
    url=('https://club.jd.com/comment/productPageComments.action?callback=fetchJSON_comment98'
         '&productId=53336639517'
         '&score=0'
         '&sortType=5'
         f'&page={i}'
         '&pageSize=10'
         '&isShadowSku=0'
         '&fold=1')

    response= requests.get(url=url)
    if response.status_code == 200:
        data=response.text
        data = data[data.find('{'):data.rfind('}')+1]
        data = json.loads(data)
    else:
        print('Request failed, status code：', response.status_code)


    data_comments = data['comments']

    for t in data_comments:
        nickname = t.get('nickname', '') 
        VIP = t.get('plusAvailable', False)  
        time_value = t.get('creationTime', '')  
        location = t.get('location', 'Unknown')  
        content = t.get('content', '') 
        item = [nickname, VIP, time_value, location, content]
        items.append(item)
```

```python
df = pd.DataFrame(items,columns=['nickname','plusAvailable','creationTime','location','content'])
df
```
